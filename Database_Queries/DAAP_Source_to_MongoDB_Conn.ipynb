```python
# The following code will convert the 1st dataset black_friday.csv to json format for further ingestion into the system.

import csv
import json

# Defining a function to convert the data:
def csv_to_json_convert(csv_file_path, json_file_path):
    # Open the CSV file from the specified location(where jupyter notebook location is configured) for reading the data
    with open(csv_file_path, 'r',encoding="utf-8") as csv_file:
        # Read the CSV file using a dictionary reader as shown below:
        csv_reader = csv.DictReader(csv_file)
        # Convert the CSV data to a list of dictionaries using the code below
        data = list(csv_reader)
    
    # Open the JSON file for writing the data and store it in the location where jupyter notebook location is configured
    with open(json_file_path, 'w') as json_file:
        # Write the JSON data to the file which you mentioned
        json.dump(data, json_file, indent=4)

# Function to call the conversion script        
csv_to_json_convert('DATASETS/black_friday.csv', 'DATASETS/black_friday.json')
```
